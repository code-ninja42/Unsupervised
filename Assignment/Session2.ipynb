Task 1:- Plot Points and Assign to Nearest Centroid 

```python
import numpy as np
import matplotlib.pyplot as plt

points = np.array([
    [2,3],
    [5,8],
    [1,2],
    [6,9],
    [7,7]
])

# Randomly choose two centroids
np.random.seed(42)
centroids = points[np.random.choice(len(points),2,replace=False)]

print("Initial Centroids:")
print(centroids)

for point in points:
    d1 = np.linalg.norm(point-centroids[0])
    d2 = np.linalg.norm(point-centroids[1])

    if d1 < d2:
        cluster = "Cluster 1"
    else:
        cluster = "Cluster 2"

    print(f"{point} -> {cluster}")

# Plot
plt.scatter(points[:,0],points[:,1],c='blue',label="Points")
plt.scatter(centroids[:,0],centroids[:,1],c='red',marker='X',s=200,label="Centroids")

for i,p in enumerate(points):
    plt.text(p[0]+0.1,p[1]+0.1,f"P{i+1}")

plt.legend()
plt.grid()
plt.show()
```

Task 2: assign_clusters() using Euclidean Distance

```python
import numpy as np

def assign_clusters(points, centroids):

    assignments=[]

    for point in points:

        distances=[np.linalg.norm(point-c) for c in centroids]

        assignments.append(np.argmin(distances))

    return assignments


points=np.array([
    [2,3],
    [5,8],
    [1,2],
    [6,9],
    [7,7]
])

centroids=np.array([
    [2,3],
    [6,9]
])

print(assign_clusters(points,centroids))
```

Task 3: Euclidean + Manhattan Distance

```python
import numpy as np

def assign_clusters(points, centroids, distance="euclidean"):

    assignments=[]

    for point in points:

        distances=[]

        for centroid in centroids:

            if distance=="euclidean":
                d=np.linalg.norm(point-centroid)

            elif distance=="manhattan":
                d=np.sum(np.abs(point-centroid))

            else:
                raise ValueError("Choose euclidean or manhattan")

            distances.append(d)

        assignments.append(np.argmin(distances))

    return assignments


points=np.array([
    [2,3],
    [5,8],
    [1,2],
    [6,9],
    [7,7]
])

centroids=np.array([
    [2,3],
    [6,9]
])

print("Euclidean:",assign_clusters(points,centroids,"euclidean"))

print("Manhattan:",assign_clusters(points,centroids,"manhattan"))
```

Task 4: Update Centroids

```python
import numpy as np

def update_centroids(points, assignments, k):

    new_centroids=[]

    for cluster in range(k):

        cluster_points=points[np.array(assignments)==cluster]

        centroid=cluster_points.mean(axis=0)

        new_centroids.append(centroid)

    return np.array(new_centroids)


points=np.array([
    [2,3],
    [5,8],
    [1,2],
    [6,9],
    [7,7]
])

assignments=[0,1,0,1,1]

print(update_centroids(points,assignments,2))
```

Task 5: Elbow Method

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample Restaurant Dataset

df=pd.DataFrame({

    "Rating":[4.8,4.5,3.9,4.2,4.9,3.8,4.1,4.7,3.6,4.0],

    "Price":[1000,900,350,450,1200,250,600,950,200,500]

})

X=df

wcss=[]

for k in range(1,7):

    model=KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    model.fit(X)

    wcss.append(model.inertia_)

plt.plot(range(1,7),wcss,marker="o")

plt.xlabel("Number of Clusters (k)")

plt.ylabel("WCSS")

plt.title("Elbow Method")

plt.grid()

plt.show()
```
