Task 1:- Apply DBSCAN

```python
import pandas as pd
from sklearn.cluster import DBSCAN

# Load Dataset
df = pd.read_csv("Mall_Customers.csv")

# Select Features
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Apply DBSCAN
dbscan = DBSCAN(eps=8, min_samples=5)
labels = dbscan.fit_predict(X)

# Number of Clusters
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

# Number of Noise Points
n_noise = list(labels).count(-1)

print("Clusters Found:", n_clusters)
print("Noise Points:", n_noise)
```

Task 2: Visualize DBSCAN Clusters

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

unique_labels = set(labels)

for label in unique_labels:

    if label == -1:
        color = "black"
        name = "Noise"
    else:
        color = plt.cm.tab10(label)
        name = f"Cluster {label}"

    cluster = X[labels == label]

    plt.scatter(
        cluster.iloc[:,0],
        cluster.iloc[:,1],
        color=color,
        label=name
    )

plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.title("DBSCAN Clustering")
plt.legend()
plt.grid(True)

plt.show()
```

Task 3: Change eps and min_samples

```python
from sklearn.cluster import DBSCAN

# Model 1
db1 = DBSCAN(eps=5, min_samples=5)
labels1 = db1.fit_predict(X)

print("Model 1")
print("Clusters:", len(set(labels1)) - (1 if -1 in labels1 else 0))
print("Noise:", list(labels1).count(-1))


# Model 2
db2 = DBSCAN(eps=10, min_samples=3)
labels2 = db2.fit_predict(X)

print("\nModel 2")
print("Clusters:", len(set(labels2)) - (1 if -1 in labels2 else 0))
print("Noise:", list(labels2).count(-1))
```

Task 4: Compare DBSCAN and K-Means

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# DBSCAN
dbscan = DBSCAN(eps=8, min_samples=5)
db_labels = dbscan.fit_predict(X)

clusters = len(set(db_labels)) - (1 if -1 in db_labels else 0)

# K-Means
kmeans = KMeans(n_clusters=clusters, random_state=42)
km_labels = kmeans.fit_predict(X)

plt.figure(figsize=(12,5))

# DBSCAN
plt.subplot(1,2,1)
plt.scatter(
    X.iloc[:,0],
    X.iloc[:,1],
    c=db_labels,
    cmap='tab10'
)
plt.title("DBSCAN")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")

# KMeans
plt.subplot(1,2,2)
plt.scatter(
    X.iloc[:,0],
    X.iloc[:,1],
    c=km_labels,
    cmap='tab10'
)
plt.title("K-Means")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")

plt.tight_layout()
plt.show()
```

Task 5 :- DBSCAN on Non-Spherical Dataset

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Generate Dataset
X, y = make_moons(
    n_samples=300,
    noise=0.08,
    random_state=42
)

# DBSCAN
dbscan = DBSCAN(
    eps=0.2,
    min_samples=5
)

labels = dbscan.fit_predict(X)

# Noise Indices
noise_indices = [i for i, label in enumerate(labels) if label == -1]

print("Noise Point Indices:")
print(noise_indices)

# Plot
plt.figure(figsize=(7,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=labels,
    cmap="tab10"
)

plt.title("DBSCAN on make_moons")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```
