```python
import pandas as pd
import numpy as np

# Dataset
orders = pd.DataFrame({
    "Order": ["O1", "O2", "O3", "O4", "O5", "O6"],
    "Delivery_Time": [20, 25, 18, 35, 40, 30],
    "Distance": [2, 3, 1, 8, 9, 6]
})

print(orders)

# Calculate Euclidean Distance Matrix
points = orders[["Delivery_Time", "Distance"]].values

distance_matrix = np.zeros((len(points), len(points)))

for i in range(len(points)):
    for j in range(len(points)):
        distance_matrix[i][j] = np.sqrt(
            (points[i][0]-points[j][0])**2 +
            (points[i][1]-points[j][1])**2
        )

distance_df = pd.DataFrame(
    distance_matrix,
    index=orders["Order"],
    columns=orders["Order"]
)

print("\nDistance Matrix")
print(distance_df.round(2))
```

Task 2: Agglomerative Hierarchical Clustering on Songs Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering

# Songs dataset
songs = pd.DataFrame({
    "Tempo":[90,95,100,105,130,135,140,145],
    "Energy":[30,35,40,45,80,85,90,95]
})

X = songs.values

# Agglomerative Clustering
model = AgglomerativeClustering(n_clusters=3)
labels = model.fit_predict(X)

print("Cluster Labels:")
print(labels)

# Dendrogram
Z = linkage(X, method="single")

plt.figure(figsize=(10,5))
dendrogram(Z)
plt.title("Songs Dendrogram (Single Linkage)")
plt.xlabel("Songs")
plt.ylabel("Distance")
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

methods = ["single", "complete", "average"]

plt.figure(figsize=(18,5))

for i, method in enumerate(methods):
    Z = linkage(X, method=method)

    plt.subplot(1,3,i+1)
    dendrogram(Z)
    plt.title(f"{method.capitalize()} Linkage")
    plt.xlabel("Songs")
    plt.ylabel("Distance")

plt.tight_layout()
plt.show()
```
