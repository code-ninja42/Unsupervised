```python
import random

users = []

for i in range(1, 11):
    user = {
        "User_ID": f"U{i}",
        "Daily_App_Opens": random.randint(1, 25),
        "Avg_Session_Duration": round(random.uniform(2, 60), 2)
    }
    users.append(user)

print(users)
```

# Supervised Learning
* Uses labeled data 
* Predicts a known target variable
* Example:- Instagram feed ranking predicts which posts a user will like 
* Algo :- Decision Tree,Random Forest , SVM

# Unsupervised Learning
* Useds Unlabeled data
* Finds hidden patterns or groups 
* Example :- Spotify groups similar songs into playlists using clustering 
* Algo :- K-Means,DBSCAN,Hierarchical Clustering
