```python
# Import libraries
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Load dataset
iris = load_iris()
X = iris.data

# Hierarchical clustering using Ward linkage
ward_linkage = linkage(X, method='ward')

# Plot dendrogram
plt.figure(figsize=(12,6))
dendrogram(ward_linkage)
plt.title("Hierarchical Clustering Dendrogram (Ward Linkage)")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```

Task 2: Cut the Dendrogram into 3 Clusters

Choose a height after looking at the dendrogram. For Iris, around 15 usually gives 3 clusters.

```python
from scipy.cluster.hierarchy import fcluster
import pandas as pd

# Cut dendrogram
clusters = fcluster(ward_linkage, t=15, criterion='distance')

# Print cluster of each sample (optional)
print(clusters)

# Count samples in each cluster
cluster_counts = pd.Series(clusters).value_counts().sort_index()

print("\nNumber of samples in each cluster:")
print(cluster_counts)
```

Task 3: Compare Ward, Single and Complete Linkage

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Create linkage matrices
ward = linkage(X, method='ward')
single = linkage(X, method='single')
complete = linkage(X, method='complete')

# Plot all three dendrograms
plt.figure(figsize=(18,5))

# Ward
plt.subplot(1,3,1)
dendrogram(ward)
plt.title("Ward Linkage")
plt.xlabel("Samples")
plt.ylabel("Distance")

# Single
plt.subplot(1,3,2)
dendrogram(single)
plt.title("Single Linkage")
plt.xlabel("Samples")
plt.ylabel("Distance")

# Complete
plt.subplot(1,3,3)
dendrogram(complete)
plt.title("Complete Linkage")
plt.xlabel("Samples")
plt.ylabel("Distance")

plt.tight_layout()
plt.show()
```

Task 4: Interpretation

In a dendrogram, clusters that merge at lower heights are more similar because the distance between them is small. Clusters that merge at higher heights are more different, indicating greater dissimilarity. Ward linkage forms compact and balanced clusters, while Single linkage often creates long chain-like clusters. Complete linkage produces tighter and more separated clusters than Single linkage.
