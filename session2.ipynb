Que 2 :define modular Python script to automate the initial data ingestion and
feature-label splitting process.

```python
import pandas as pd

def load_data(file_path):
    df = pd.read_csv(file_path)
    print("Dataset loaded successfully!")
    return df

def dataset_info(df):
    print("Shape:", df.shape)
    print("Columns:")
    print(df.columns)
    print("\nFirst 5 Rows:")
    print(df.head())

def split_feature_label(df, target_column):
    X = df.drop(columns=[target_column])
    y = df[target_column]
    return X, y


# Main
file_path = "IBM_HR.csv"

df = load_data(file_path)
dataset_info(df)
X, y = split_feature_label(df, "Attrition")
```
