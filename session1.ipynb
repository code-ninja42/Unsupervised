1.Python Script to Generate a Random Dataset of 10 Users

```python
import random
import random

users = []

for i in range(1, 11):
    user = {
        "User_ID": f"U{i}",
        "Daily_App_Opens": random.randint(1, 30),
        "Avg_Session_Duration": round(random.uniform(2, 60), 2)  # minutes
    }
    users.append(user)

print(users)
```

2. Comparison of Supervised vs Unsupervised Learning

* Supervised Learning
1.Training Data - Uses labeled data (input + correct output)
2.Goal - Predict a target value 
3.Output - Classification or Regression
4.Real-world Example - Instagram feed ranking predicts which posts users will like
5.Algorithms - Decision Tree,Random Forest,SVM

*Unsupervised Learning
1.Training Data - Uses unlabeled data
2.Goal - Discover hidden patterns or groups
3.Output - Cluster or Assiociation
4.Real-worls Example - Spotify groups similar songs into playlists based on listening patterns
5.Algorithms - K-Means,DBSCAN,Hierarchical Clustering

```python
import pandas as pd
from sklearn.cluster import KMeans

# ----------------------------
# Create Dataset
# ----------------------------
df = pd.DataFrame({
    "Product": [
        "Phone A", "Phone B",
        "Laptop A", "Laptop B",
        "Earbuds A", "Earbuds B",
        "Smartwatch A", "Smartwatch B"
    ],
    "Price": [800, 750, 1200, 1150, 50, 60, 250, 230],
    "Reviews": [5000, 4500, 3000, 2800, 700, 900, 1800, 1700]
})

print("Original Dataset")
print(df)

# ----------------------------
# Select Features
# ----------------------------
X = df[["Price", "Reviews"]]

# ----------------------------
# Create KMeans Model
# ----------------------------
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

# ----------------------------
# Fit Model
# ----------------------------
df["Cluster"] = kmeans.fit_predict(X)

# ----------------------------
# Cluster Centers
# ----------------------------
print("\nCluster Centers")
print(kmeans.cluster_centers_)

# ----------------------------
# Final Output
# ----------------------------
print("\nProducts with Cluster Labels")
print(df.sort_values("Cluster"))
```

```python
import pandas as pd
from sklearn.mixture import GaussianMixture

# ----------------------------
# Create Dataset
# ----------------------------
df = pd.DataFrame({
    "Product": [
        "Phone A", "Phone B",
        "Laptop A", "Laptop B",
        "Earbuds A", "Earbuds B",
        "Smartwatch A", "Smartwatch B"
    ],
    "Price": [800, 750, 1200, 1150, 50, 60, 250, 230],
    "Reviews": [5000, 4500, 3000, 2800, 700, 900, 1800, 1700]
})

print("Original Dataset")
print(df)

# ----------------------------
# Features
# ----------------------------
X = df[["Price", "Reviews"]]

# ----------------------------
# Create GMM Model
# ----------------------------
gmm = GaussianMixture(
    n_components=3,
    random_state=42
)

# ----------------------------
# Train Model
# ----------------------------
gmm.fit(X)

# ----------------------------
# Predict Cluster
# ----------------------------
df["Cluster"] = gmm.predict(X)

# ----------------------------
# Probability of Membership
# ----------------------------
probabilities = gmm.predict_proba(X)

print("\nCluster Membership Probability")
print(probabilities)

print("\nFinal Clusters")
print(df.sort_values("Cluster"))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    df["Price"],
    df["Reviews"],
    c=df["Cluster"],
    cmap="rainbow",
    s=120
)

for i in range(len(df)):
    plt.text(
        df["Price"][i] + 10,
        df["Reviews"][i],
        df["Product"][i]
    )

plt.xlabel("Price")
plt.ylabel("Reviews")
plt.title("Gaussian Mixture Model Clustering")
plt.grid(True)
plt.show()
```
