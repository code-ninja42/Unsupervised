# Project Title 

Predictive Exit Interview System

# Business Problem 

Employee attrition increases recruitment costs, reduces productivity, and causes knowledge loss. The organization needs a predictive system to identify employees who are at high risk of leaving before they resign.

# Objective

Develop a machine learning model that predicts whether an employee is likely to leave the organization (Attrition = Yes/No)

# Dataset

IBM HR Analytics Employee Attrition Dataset

Source: Kaggle
Records: 1470
Features: 35
Target Variable: Attrition
